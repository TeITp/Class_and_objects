```python
class MyName:
    """Оновлений клас для Лабораторної роботи №1"""
    total_names = 0

    def __init__(self, name=None) -> None:
        self.name = name if name is not None else self.anonymous_user().name

        # ЗАВДАННЯ: Перевірка на літери (валідація)
        # Якщо ім'я містить щось крім букв (наприклад цифри), піднімаємо помилку
        if not self.name.isalpha():
             raise ValueError("Ім'я може містити лише літери!")

        # ЗАВДАННЯ: Ім'я завжди з великої літери
        self.name = self.name.capitalize()

        MyName.total_names += 1
        self.my_id = self.total_names

    @property
    def whoami(self) -> str:
        return f"My name is {self.name}"

    @property
    def my_email(self) -> str:
        return self.create_email()

    # ЗАВДАННЯ: Властивість full_name
    @property
    def full_name(self) -> str:
        return f"User #{self.my_id}: {self.name} ({self.my_email})"

    # ЗАВДАННЯ: Модифікація create_email (можна міняти домен)
    def create_email(self, domain="itcollege.lviv.ua") -> str:
        return f"{self.name}@{domain}"

    # ЗАВДАННЯ: Підрахунок букв
    def count_letters(self) -> int:
        return len(self.name)

    # ЗАВДАННЯ: Збереження у файл
    def save_to_file(self, filename="users.txt"):
        with open(filename, "a", encoding="utf-8") as file:
            file.write(self.full_name + "\n")

    @classmethod
    def anonymous_user(cls):
        return cls("Anonymous")

    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        return f"You say: {message}"

# --- Блок виконання та перевірки ---
print("Розпочинаємо роботу модифікованої програми!")

# ЗАВДАННЯ: Додати своє ім'я в список
# "R2D2" додано спеціально, щоб показати роботу помилки ValueError
names = ("bohdan", "Marta", None, "Student", "R2D2")

for name in names:
    try:
        # Створюємо об'єкт
        me = MyName(name)

        print(f"\nCreated: {me.full_name}")
        print(f"Letters count: {me.count_letters()}")

        # ЗАВДАННЯ: Змінити текст привітання
        print(me.say_hello("Привіт, я успішно виконав завдання!"))

        # ЗАВДАННЯ: Перевірка зміни емейлу
        print(f"Custom Email: {me.create_email('gmail.com')}")

        # ЗАВДАННЯ: Збереження
        me.save_to_file()
        print("-> Saved to file.")

    except ValueError as e:
        print(f"\n⛔ ПОМИЛКА для імені '{name}': {e}")

print(f"\nВсього імен (лічильник класу): {MyName.total_names}")
```
